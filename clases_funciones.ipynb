```python
class Unificado():

    def __init__(self,dataframe=pd.DataFrame()):
        self.dataframe = dataframe

    def union_dataframes(self,*args):
        '''
        Esta función recibe una tupla con los diferentes dataframes que queremos unir. Une los dataframes por columnas y 
        devuelve un dataframe con todo unido.
        '''
        lista = list(args)
        union = pd.concat(lista,axis=1)
        self.dataframe = union
        return union

    def parametros(self):
        """Esta función muestra un estudio general de un dataframe.

        Args:
            dataframe (dataframe): el dataframe de cual queremos hacer su estudio.
        return: a pesar de que esta función no devuelve nada, mustra en pantalla todo el estudio realizado.
        """
        print(f'El tamaño del DataFrame es {self.dataframe.size}')
        print('------------------------------------------------------------------------')
        print(f'El número de filas y columnas del DataFrame es {self.dataframe.shape}')
        print('------------------------------------------------------------------------')
        print('La descripción estadística del DataFrame es:')
        display(self.dataframe.describe())
        print('------------------------------------------------------------------------')
        print('La información general del DataFrame es:')
        display(self.dataframe.info())
        print('------------------------------------------------------------------------')
        print(f'Las columnas de este DataFrame son {self.dataframe.columns}')
        print('------------------------------------------------------------------------')
        print(f'El tipo de dato de las columnas de este DataFrame es {self.dataframe.dtypes}')
        print('------------------------------------------------------------------------')
        print(f'Los nulos de este DataFrame son {self.dataframe.isnull().sum()}')
    
    def union(self,*args):
        '''
        Esta función recibe una tupla con los elementos de varias columnas. Une los datos por comas y tiene en cuenta los nulos, 
        devuelve un str de los elementos unidos por comas.'''
        lista_str = [str(x) for x in args]
        lista_final = [i for i in lista_str if i.lower() != 'nan']
        texto = ','.join(lista_final)
        if len(lista_final) == 0 :
            return np.nan
        else:   
            return texto

    def eliminar_col(self, *args):
        '''
        Esta función recibe una tupla con los elementos de varias columnas. Elimina todas
        las columnas que haya en la tupla.'''
        lista = list(args)
        self.dataframe.drop(lista, axis = 1, inplace = True)
    
    def guardar_archivo(self,ruta,formato):
        '''
        Esta función guarda el archivo en formato especificado.'''
        if formato == 'csv':
            self.dataframe.to_csv(ruta)
        elif formato == 'pkl':
            self.dataframe.to_pickle(ruta)

```
