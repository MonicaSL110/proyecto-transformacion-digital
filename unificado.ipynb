```python
def union(*args):

    lista_str = [str(x) for x in args]
    lista_final = [i for i in lista_str if i.lower() != 'nan']
    texto = ', '.join(lista_final)
    if len(lista_final) == 0 :
        return np.nan
    else:   
        return texto
```
