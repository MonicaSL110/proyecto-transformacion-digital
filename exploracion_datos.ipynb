### Creación de una función de propiedades de DataFrames

```python
def parametros(dataframe): 

     print(f"El tamaño del DataFrame es {dataframe.size}")
     print("------------------------------------------------------------------------")
     print(f"El número de filas y columnas del DataFrame es {dataframe.shape}")
     print("------------------------------------------------------------------------")
     print(f"La descripción estadística del DataFrame es {dataframe.describe()}")
     print("------------------------------------------------------------------------")
     print(f"La información general del DataFrame es {dataframe.info}")
     print("------------------------------------------------------------------------")
     print(f"Las columnas de este DataFrame son {dataframe.columns}")
     print("------------------------------------------------------------------------")
     print(f"El tipo de dato de las columnas de este DataFrame es {dataframe.dtypes}")

# Creamos una función para poder explorar los datos de cada DataFrame. 
```

```python
parametros(data_csv1)
```

```python
parametros(data_csv2)
```

```python
parametros(data_txt)
```

```python
parametros(data_xml)
```
