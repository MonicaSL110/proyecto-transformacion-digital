### Apertura del archivo .txt y creación de un DataFrame



```python
# Reutilizamos el código de automatización de apertura y procesado de datos de los archivos txt para que nos devuelve una lista de listas. 

class Archivo_Txt:
    def __init__(self,ruta):
        self.ruta = ruta
    
    def apertura_txt(self, archivo_txt):
        with open(self.ruta+archivo_txt,'r') as file: #self.ruta/arch                                                     ivo_txt
            file_source = file.readlines()
        return file_source
    
    def procesado_txt(self, archivo_txt):
        lista_source = [archivo_txt]
        file_limpio = []

        for lista in lista_source:
            for campo in lista:
                file_limpio.append(campo.replace("\n", "").replace("null", "NULL").replace("<","under").replace("   "," ").replace("  "," ").replace(" ,",",").strip())
        
        return file_limpio
    
    def creacion_tabla_txt(self, archivo_txt):
        lista_tupla = []
        for linea in archivo_txt:
            lista_tupla.append((linea.split(";")))
            
        return lista_tupla
```

```python
data_txt = Archivo_Txt("datos/")
```

```python
apertura_txt = data_txt.apertura_txt("/data_txt_clean.txt")
```

```python
procesado_txt = data_txt.procesado_txt(apertura_txt)
```

```python
crear_tabla = data_txt.creacion_tabla_txt(procesado_txt)
crear_tabla
```

```python
lista_datos_txt = []

for valores in crear_tabla[1:]:
    resultado = dict(zip(crear_tabla[0], valores))
    lista_datos_txt.append(resultado)


print(lista_datos_txt[2])

# Creamos un diccionario para poder pasar los datos a un DataFrame. 
```

```python
datos_txt = pd.DataFrame(lista_datos_txt)
datos_txt.head(1)

#Creamos el DataFrame con los datos del archivo .txt una vez se ha convertido en un diccionario. 
```

```python
datos_txt.set_index(["index"], inplace=True)
datos_txt.head(1)
# Indicamos que el el index de este DataFrame sea la columna "index".
```

### Apertura del archivo .xml y creación de un DataFrame

```python
# Reutilizamos el código de automatización de apertura y procesado de datos de los archivos txt para que nos devuelve una lista de diccionarios.

class Archivo_Xml:
    def __init__(self,ruta):
        self.ruta = ruta
        
    def apertura_xml(self,archivo_xml):
    
        tree = ET.parse(self.ruta+archivo_xml)
        root = tree.getroot()

        xml_list = []
        for child in root:
            xml_dict = {}
            for subchild in child:
                xml_dict[subchild.tag] = subchild.text
            xml_list.append(xml_dict)
        
        return xml_list

    def procesado_xml(self,archivo_xml):
        for fila in archivo_xml:
            if fila['gender'] == '0':
                fila['gender'] = 'Man'
            elif fila['gender'] == '1':
                fila['gender'] = 'Woman'
            elif fila['gender'] == '2':
                fila['gender'] = 'Nonbinary'
            elif fila['gender'] == '3':
                fila['gender'] = 'Prefer not to say'
            elif fila['gender'] == '4':
                fila['gender'] = 'Prefer to self-describe'

        for fila in archivo_xml:
            if 'level_0' in fila:
                fila.pop('level_0')
            else:
                pass
            
        return archivo_xml
```

```python
archivo_xml = Archivo_Xml("datos/")
```

```python
apertura_xml = archivo_xml.apertura_xml("/data_xml_clean.xml")
```

```python
procesado_xml = archivo_xml.procesado_xml(apertura_xml)
procesado_xml[1]
```

```python
datos_xml = pd.DataFrame(procesado_xml)
datos_xml.head(1)

#Creamos el DataFrame con los datos del archivo .xml
```
